```python
import  cv2,os
import tkinter as tk
from tkinter import filedialog
from tkinter import *
from PIL import ImageTk, Image
from PIL import ImageOps
import numpy as np
from keras.models import load_model
import tensorflow as tf




model = load_model(r'C:\Users\Mahmudul Hasan\Desktop\System project\BatchNorm250.model\BatchNorm250.model')

# dictionary to label all traffic signs class.
classes = {1: 'Empty',
           2:'High',
           3:'Low',
           4:'Medium',
           5:'Traffic Jam'
           }


def open_window():
    def show_loss_image():
        try:
            lossed = Image.open(r'C:\Users\Mahmudul Hasan\Desktop\System project\PAPER IMAGES\4(b).png')
            lossed.thumbnail(((top.winfo_width() / 2.25), (top.winfo_height() / 2.25)))
            im = ImageTk.PhotoImage(lossed)

            graph_image.configure(image=im)
            graph_image.image = im
            #label.configure(text='')
            show.configure(foreground='black',text="Loss vs epoch Graph")

        except:
            pass
    def show_accuracy_image():
        try:
            ac = Image.open(r'C:\Users\Mahmudul Hasan\Desktop\System project\PAPER IMAGES\5.png')
            ac.thumbnail(((top.winfo_width() / 2.25), (top.winfo_height() / 2.25)))
            im = ImageTk.PhotoImage(ac)

            graph_image.configure(image=im)
            graph_image.image = im
            #label.configure(text='')
            show.configure(foreground='black',text="Accuracy vs epoch Graph")
        except:
            pass
    def show_confusion_image():
        try:
            con_matrix = Image.open(r'C:\Users\Mahmudul Hasan\Desktop\System project\PAPER IMAGES\6.png')
            con_matrix.thumbnail(((top.winfo_width() / 2.25), (top.winfo_height() / 2.25)))
            im = ImageTk.PhotoImage(con_matrix)

            graph_image.configure(image=im)
            graph_image.image = im
            #label.configure(text='')
            show.configure(foreground='black',text="Confusion Matrix")
        except:
            pass
    def show_roc_image():
        try:
            roc_graph = Image.open(r'C:\Users\Mahmudul Hasan\Desktop\System project\PAPER IMAGES\7.png')
            roc_graph.thumbnail(((top.winfo_width() / 2.25), (top.winfo_height() / 2.25)))
            im = ImageTk.PhotoImage(roc_graph)

            graph_image.configure(image=im)
            graph_image.image = im
            #label.configure(text='')
            show.configure(foreground='black',text="ROC curve")
        except:
            pass
    def show_precision_recall_image():
        try:
            pr_graph = Image.open(r'C:\Users\Mahmudul Hasan\Desktop\System project\PAPER IMAGES\8.png')
            pr_graph.thumbnail(((top.winfo_width() / 2.25), (top.winfo_height() / 2.25)))
            im = ImageTk.PhotoImage(pr_graph)

            graph_image.configure(image=im)
            graph_image.image = im
            #label.configure(text='')
            show.configure(foreground='black',text="Precision-recall curve")
        except:
            pass
    new_window=Toplevel(top)
    new_window.title("Traffic Density Estimation")
    new_window.geometry("1200x900")
    new_window.configure(background='#242526')
    heading = Label(new_window, text="Traffic Density Estimation", pady=20, font=('Times New Roman', 30, 'bold'))
    heading.configure(background='#242526', foreground='#0084ff')
    heading.pack()
    
    show = Label(new_window, background='#CDCDCD', font=('arial', 21, 'bold'))
    show.pack()
    
    loss = Button(new_window, text="Loss vs epoch Graph", command=show_loss_image, padx=10, pady=5)
    loss.configure(background='#3e4042', foreground='white', font=('arial', 10, 'bold'),height = 1, width = 20)
    loss.place(relx=0.1, rely=0.3)

    accuracy = Button(new_window, text="Accuracy vs epoch Graph",command=show_accuracy_image, padx=10, pady=5)
    accuracy.configure(background='#3e4042', foreground='white', font=('arial', 10, 'bold'),height = 1, width = 20)
    accuracy.place(relx=0.1, rely=0.4)

    confusion_matrix = Button(new_window, text="Confusion Matrix",command=show_confusion_image, padx=10, pady=5)
    confusion_matrix.configure(background='#3e4042', foreground='white', font=('arial', 10, 'bold'),height = 1, width = 20)
    confusion_matrix.place(relx=0.1, rely=0.5)

    roc = Button(new_window, text="ROC Curve",command=show_roc_image, padx=10, pady=5)
    roc.configure(background='#3e4042', foreground='white', font=('arial', 10, 'bold'),height = 1, width = 20)
    roc.place(relx=0.1, rely=0.6)

    pr = Button(new_window, text="Precision Recall Graph",command=show_precision_recall_image, padx=10, pady=5)
    pr.configure(background='#3e4042', foreground='white', font=('arial', 10, 'bold'),height = 1, width = 20)
    pr.place(relx=0.1, rely=0.7)

    label2 = Label(new_window,background='#242526')
    label2.pack(side=BOTTOM, expand=True)
    
    label4 = Label(new_window,background='#242526')
    label4.pack(side=BOTTOM, expand=True)
    
    graph_image = Label(new_window,background='#242526')
    graph_image.pack(side=BOTTOM, expand=True)
    
    
    label5 = Label(new_window,background='#242526')
    label5.pack(side=BOTTOM, expand=True)

def classify(file_path):
    image = Image.open(file_path)
    image = cv2.imread(file_path,cv2.IMREAD_GRAYSCALE) 
    image = cv2.resize(image,(200,200)) 
    image = image.reshape(1, 200, 200,1).astype('float32')
#	image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
#	ret,image=cv2.threshold(image,120,255,cv2.THRESH_BINARY)
    image = np.array(image)/255
    predictions = model.predict(x=image, verbose=0)
    np.round(predictions)
    pred=np.argmax(predictions, axis=-1)[0]
    print(image.shape)
   # pred = model.predict_classes([image])[0]
    density = classes[pred + 1]
    print(density)
    label.configure(foreground='black',text=density)


def show_classify_button(file_path):
    classify_b = Button(top, text="Classify Traffic Density", command=lambda: classify(file_path), padx=10, pady=5)
    classify_b.configure(background='#3e4042', foreground='white', font=('arial', 10, 'bold'),height = 1, width = 20)
    classify_b.place(relx=0.79, rely=0.46)
    #classify_b.after(5000, classify_b.destroy)


def upload_image():
	try:
		file_path = filedialog.askopenfilename()
		uploaded = Image.open(file_path)
		uploaded.thumbnail(((top.winfo_width() / 2.25), (top.winfo_height() / 2.25)))
		im = ImageTk.PhotoImage(uploaded)

		density_image.configure(image=im)
		density_image.image = im
		label.configure(text='')
		show_classify_button(file_path)
	except:
		pass

# initialise GUI
top = tk.Tk()
top.geometry('1500x900')
top.title("Traffic Density Estimation ")
top.configure(background='#242526')

#set heading
heading = Label(top, text="Traffic Density Estimation", pady=20, font=('Times New Roman', 30, 'bold'))
heading.configure(background='#242526', foreground='#0084ff')
heading.pack()

upload = Button(top, text="Upload a traffic image", command=upload_image, padx=10, pady=5)
upload.configure(background='#3e4042', foreground='white', font=('arial', 10, 'bold'),height = 1, width = 20)
upload.pack(side=BOTTOM, pady=60)

graph = Button(top, text="Evaluate the Model", command=open_window, padx=10, pady=5)
graph.configure(background='#3e4042', foreground='white', font=('arial', 10, 'bold'),height = 1, width = 20)
graph.place(relx=0.1, rely=0.3)

density_image = Label(top,background='#242526')
density_image.pack(side=BOTTOM, expand=True)

label3 = Label(top,background='#242526')
label3.pack(side=BOTTOM, expand=True)


label = Label(top, background='#CDCDCD', font=('arial', 21, 'bold'))
label.place(relx=0.593,rely=0.15)


label1 = Label(top, background='#CDCDCD', font=('arial', 21, 'bold'))
label1.configure(foreground='black',text="Estimated Traffic Density: ")
label1.place(relx=0.33,rely=0.15)

top.mainloop()
```
